Problem 1 : Social Media -Post Engagement Analyzer


```python
#function to analyze the engagement on a social media post.
def analyzer(likes):
  total_likes = 0
  for count in likes:
    total_likes += count
  if total_likes >= 1000:
    print(f"Total Likes : {total_likes}")
    print("Post Status : Viral Post")
  else :
    print(f"Total Likes : {total_likes}")
    print("Post Status : Normal Engagement")
likes = list(map(int,input().split(",")))  #Input
analyzer(likes)       # function calling.
```

Problem 3: Agriculture-Rainfall adequacy checker


```python
#function to analyze rainfall data
def rainfall(seasonal,required):
  avg=sum(seasonal)//len(seasonal)
  if avg >= required :
    print(f"Average Rainfall :{avg}")
    print("Rainfall Status : Adequate Rainfall")
  else :
    print(f"Average Rainfall :{avg}")
    print("Rainfall Status : Inadequate Rainfall")
seasonal = list(map(int,input().split(",")))
required=int(input())
rainfall(seasonal,required)
```

Problem 4 : Social Media-Duplicate acoount Detection


```python
def detection(uname):
  s=set(uname)
  if len(s)==len(uname):
    print(f"Duplicate Accounts Found :No")
  else:
    print("Duplicate Accounts Found :Yes")
uname = list(map(str,input().split()))
detection(uname)
```

Problem 6 :Agriculture -Premium Crop Price Filter


```python
def filter(crop):
  premium_crop =[]
  regular_crop =[]
  for price in crop:
    if price > 2000:
      premium_crop.append(price)
    else:
      regular_crop.append(price)
  print(f"Premium Crops :{premium_crop}")
crop =list(map(int,input().split(",")))
filter(crop)
```

Problem 7: System Monitoring-Application Health Checker


```python
def checker(errors):
  if errors == 0 :
    print(f"System Status :Critical Issues")
  elif errors <= 5 :
    print("System Status :Minor Issues")
  elif errors >5 :
    print("Critical Issues")
errors =int(input())
print(f"Error Count : {errors}")
checker(errors)
```

Problem 8 : Banking- Daily Transaction Limit Checker


```python
def limit(amount,daily_limit):
  if amount <= daily_limit:
    print("Transaction Status : Approved")
  else:
    print("Transaction Status : Rejected")
amount=int(input())
daily_limit = int(input())
print(f"Transaction Amount :{amount}")
limit(amount,daily_limit)
```

Problem 9 : E-Learning

```python
def eligible(percentage):
  if percentage >=75:
    print("Exam Eligibilty : Eligible")
  elif percentage <75:
    print("Exam Eligibilty : Not Eligible")
attendance=list(map(str,input().split(",")))
persentee=attendance.count("P")
percentage=(persentee)/len(attendance)*100
print(f"Attendance Percentage :{percentage}")
eligible(percentage)
```

Problem 10:Smart Electricity Bill Analyzer

```python
def electricity(units):
  if units <=100 :
    bill = units*3
  elif units>100 and units <=200:
    bill=units*5
  elif units>200:
    bill=units*7
  print(f"Total Bill Amount :{bill}")
  if bill <500:
    print("Low Usage")
  elif bill >=500 and bill <=1500:
    print("Moderate Usage")
  elif bill >1500:
    print("High Usage")
units = int(input())
print(f"Units Consumed :{units}")
electricity(units)


```
