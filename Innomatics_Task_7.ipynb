Problem 1:Smart Light Controller.


```python
class SmartLight:
    def __init__(self, name, action):
        self.name = name
        self.is_on = action.strip().upper() == "ON"

    def get_status(self):
        status_str = "ON" if self.is_on else "OFF"
        return f"{self.name} is {status_str}"
light=input("Light Name :")
action=input("Action :")
result=SmartLight(light,action)
print(result.get_status())
```

Problem 2:Employee ID CArd System

```python
class Employee:
  def __init__(self,name,id,dept):
    self.name = name
    self.id=id
    self.dept=dept
  def display(self):
    print("Employee Id card")
    print(f"Name :{self.name}")
    print(f"ID :{self.id}")
    print(f"Department :{self.dept}")
name=input("Employee Name:")
id=input("Employee ID:")
dept=input("Department:")
details=Employee(name,id,dept)
print(details.display())
```

Problem 3:Mobile Contact Record

```python
class store:
  def __init__(self,name,number):
    self.name=name
    self.number=number
  def display_contact(self):
    print("Contact Saved")
    print(f"Name :{self.name}")
    print(f"Phone : {self.number}")
name=input("Contact NAme :")
number=int(input("Phone Number:"))
contact=store(name,number)
print(contact.display_contact())
```

Problem 4 :Product Price Tag Generator


```python
class Product:
  def __init__(self,name,price):
    self.name = name
    self.price=price
  def display(self):
    print(f"Product : {self.name}")
    print(f"Price : {self.price}")
name=input("Product NAme :")
price=int(input("Price"))
store=Product(name,price)
print(store.display())
```

Problem 5 :Movie Rating Display System.

```python
class Movie:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating
  def display(self):
    print(f"Movie :{self.name}")
    print(f"Rating :{self.rating}")
name=input("Movie Name :")
rating=float(input("Rating :"))
details=Movie(name,rating)
print(details.display())
```
