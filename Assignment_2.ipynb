```python
name=input("enter a string")
char_count = {}
for char in name:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

```python
attendance = ["P", "P", "A", "P","A","P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python

prices = [1450, 120, 999, 1500, 3000]
count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
employees = {
    "tillu": 98000,
    "Anita": 69000,
    "santhosh": 72000
}
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```
